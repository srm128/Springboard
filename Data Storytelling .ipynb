```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

```

```python
df1 = pd.read_csv(r'starbucks_drinkMenu_expanded.csv')
df1.head()
```

```python
df2 = pd.read_csv(r'starbucks-menu-nutrition-drinks.csv')
df2.head()
```

```python
df3 = pd.read_csv(r'starbucks-menu-nutrition-food2.csv')
df3.head()
```

```python
#checking unique values for df1 column "Beverage_prep" 
unique_beverage_prep = df1["Beverage_prep"].unique()
print(unique_beverage_prep)
```

```python
#Since df2 only contains Tall drinks (or equivalent), for df1 we are only keeping Tall drinks. 
df1 = df1[df1["Beverage_prep"] == "Tall"]
df1.head()
```

```python
# removing leading and trailing spaces. 
df1.columns = df1.columns.str.strip()
print(df1.columns)
```

```python
# df2 and df3 only list the following: Calories, Fat (g), Carb. (g), Fiber (g), Protein (g).
# We will have df1 match
df1 = df1[["Beverage", "Calories", "Total Fat (g)", "Total Carbohydrates (g)", "Dietary Fibre (g)", "Protein (g)"]]
df1 = df1.rename(columns={"Beverage": "Product", "Total Fat (g)": "Fat (g)", "Total Carbohydrates (g)": "Carb. (g)", "Dietary Fibre (g)": "Fiber (g)"})
df1.head()
```

```python
df2.head()
```

```python
# Rename column to Beverage
df2 = df2.rename(columns={"Unnamed: 0": "Product", "Protein": "Protein (g)"})
df2.head()
```

```python
#dropping Sodium from df2
df2 = df2.drop("Sodium", axis=1)
```

```python
df3 = df3.rename(columns={"Unnamed: 0": "Product"})
df3.head()
```

```python
df1['Type'] = 'Beverage'
df2['Type'] = 'Beverage'
df3['Type'] = 'Food'
```

```python
df1.head()
```

```python
df = pd.concat([df1, df2, df3], ignore_index=True)
```

```python
df.head()
```

```python
# drop rows with NaN values from df
df = df.dropna()
```

```python
# check if df contain NaN values
print(df.isnull().any())
```

```python
# check unique values in all rows of df
for column in df.columns:
    unique_values = df[column].unique()
    print(f"Unique values in {column}: {unique_values}")
```

```python
# view rows that contain "-" in specific columns
cols = ['Calories', 'Fat (g)', 'Carb. (g)', 'Fiber (g)', 'Protein (g)']
mask = df[cols].apply(lambda x: x.str.contains("-")).any(axis=1)
rows_with_dash = df[mask]
rows_with_dash.head(10)
```

```python
# remove rows that contain "-" in specific columns
cols = ['Calories', 'Fat (g)', 'Carb. (g)', 'Fiber (g)', 'Protein (g)']
mask = df[cols].apply(lambda x: x.str.contains("-")).any(axis=1)
df = df[~mask]
```

```python
# set the colors for each nutrient
colors = {'Fat (g)': 'red', 'Carb. (g)': 'green', 'Fiber (g)': 'blue', 'Protein (g)': 'purple'}

# set the marker shapes for each type
markers = {'Beverage': '^', 'Food': 'o'}

# create a scatterplot for each nutrient
for nutrient_name, nutrient_data in df[['Fat (g)', 'Carb. (g)', 'Fiber (g)', 'Protein (g)', 'Type', 'Calories']].dropna().groupby('Type'):
    plt.scatter(nutrient_data['Fat (g)'], nutrient_data['Calories'], c=colors['Fat (g)'], marker=markers[nutrient_name], label='Fat')
    plt.scatter(nutrient_data['Carb. (g)'], nutrient_data['Calories'], c=colors['Carb. (g)'], marker=markers[nutrient_name], label='Carb.')
    plt.scatter(nutrient_data['Fiber (g)'], nutrient_data['Calories'], c=colors['Fiber (g)'], marker=markers[nutrient_name], label='Fiber')
    plt.scatter(nutrient_data['Protein (g)'], nutrient_data['Calories'], c=colors['Protein (g)'], marker=markers[nutrient_name], label='Protein')

# set the axis labels and legend
plt.xlabel('Nutrient (g)')
plt.ylabel('Calories')
plt.legend()

# show the plot
plt.show()
```

```python
# set the colors for each type
colors = {'Beverage': 'red', 'Food': 'blue'}

# set the marker shapes for each nutrient
markers = {'Fat (g)': 'o', 'Carb. (g)': 's', 'Fiber (g)': '^', 'Protein (g)': 'D'}

# loop through each nutrient
for nutrient in ['Fat (g)', 'Carb. (g)', 'Fiber (g)', 'Protein (g)']:
    # create a scatterplot for each type
    for type_name, type_data in df.groupby('Type'):
        plt.scatter(type_data[nutrient], type_data['Calories'], c=colors[type_name], marker=markers[nutrient], label=type_name)

    # set the axis labels and legend
    plt.xlabel(nutrient)
    plt.ylabel('Calories')
    plt.legend()

    # show the plot
    plt.show()
```

```python
# count the number of missing values in each column
missing_values = df.isnull().sum()

# print the result
print(missing_values)
```

```python
# Checking unique values

print(df['Calories'].unique())
print(df['Fat (g)'].unique())
print(df['Protein (g)'].unique())
print(df['Carb. (g)'].unique())
print(df['Fiber (g)'].unique())
```

```python
# Need to make float
df['Calories'] = df['Calories'].astype(float)
df['Fat (g)'] = df['Fat (g)'].astype(float)
df['Protein (g)'] = df['Protein (g)'].astype(float)
df['Carb. (g)'] = df['Carb. (g)'].astype(float)
df['Fiber (g)'] = df['Fiber (g)'].astype(float)
```

```python
nutrient_cols = ['Calories', 'Fat (g)', 'Carb. (g)', 'Fiber (g)', 'Protein (g)']
df.groupby('Type')[nutrient_cols].mean()
```

```python
# calculate the mean calories for each type
mean_calories = df.groupby('Type')['Calories'].mean()

# create a bar plot
plt.bar(mean_calories.index, mean_calories)

# set the axis labels and title
plt.xlabel('Type')
plt.ylabel('Mean Calories')
plt.title('Mean Calories per Type')

# show the plot
plt.show()
```

```python
# checking least and most caloric foods
# create a new dataframe for each type
food_df = df[df['Type'] == 'Food']
beverage_df = df[df['Type'] == 'Beverage']

# find the top five most caloric items for each type
food_top5 = food_df.nlargest(5, 'Calories')
beverage_top5 = beverage_df.nlargest(5, 'Calories')

# print the results
print("Top five most caloric foods:")
print(food_top5[['Product', 'Calories']])

print("\nTop five most caloric beverages:")
print(beverage_top5[['Product', 'Calories']])
```

```python
# find the top five least caloric items for each type
food_bottom5 = food_df.nsmallest(5, 'Calories')
beverage_bottom5 = beverage_df.nsmallest(5, 'Calories')

# print the results
print("Top five least caloric foods:")
print(food_bottom5[['Product', 'Calories']])

print("\nTop five least caloric beverages:")
print(beverage_bottom5[['Product', 'Calories']])
```

```python
# calculate mean nutrients for each type
type_means = df.groupby('Type')[['Fat (g)', 'Carb. (g)', 'Protein (g)']].mean()

# create a list of labels and colors
labels = ['Fat (g)', 'Carb. (g)', 'Protein (g)']
colors = ['#FFC300', '#FF5733', '#900C3F']

# create a figure with two subplots
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# create a pie chart for each type
for i, type_name in enumerate(type_means.index):
    # calculate the values for the pie chart
    values = type_means.loc[type_name].values
    values = values / np.sum(values) * 100
    
    # create the pie chart
    axs[i].pie(values, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
    axs[i].axis('equal')
    axs[i].set_title(type_name)

# set the title of the figure
fig.suptitle('Mean Nutrients for Food and Beverage')

# show the plot
plt.show()
```

The analysis of the Starbucks menu revealed some interesting insights. Firstly, we found that, on average, food items have more calories than beverage items. This is not surprising, given that food items generally contain more fat, protein, and carbohydrates than beverages.

Looking at the mean nutrient values for both types, we see that food items contain significantly more fat, carbohydrates, and protein than beverage items. This is likely due to the fact that food items are more substantial and provide more energy than beverages. It appears that while food items have higher absolute values of carbohydrates, beverages have a higher proportion of their calories coming from carbohydrates. This is likely due to the fact that many of the beverage items contain added sugars, such as those found in flavored syrups and sauces. However, it is important to note that both food and beverage items are relatively carb-heavy, which can be a concern for individuals looking to manage their carbohydrate intake.

In terms of the most and least caloric items, we found that the top five most caloric items were all food items, with the highest-calorie item being the Lentils & Vegetable Protein Bowl with Brown Rice. On the other hand, the top five least caloric items included both food and beverage items, with Tazo Tea being the least caloric beverage item and the Frappuccino Cookie Straw being the least caloric food item.

Overall, this analysis provides valuable insights into the nutritional content of Starbucks menu items and can be useful for individuals looking to make informed choices about what they consume.
